```python
#Etapa 01 A


nomes = ["Tibério", "Renato", "Paulo", "Julio", "Carlos", "Lauro", "Jeferson", "Flávio", "Adjalma", "Adriano", "Michel", "Denes"]

idades = [34, 29, 29, 26, 41, 26, 27, 26, 37, 28, 41, 47]

salarios = [20000.00, 30000.00, 30000.00, 25000.00, 100000.00, 20000.00, 15000.00, 15000.00, 25000.00, 20000.00, 15000.00, 25000.00]


for n, i in zip(nomes, idades):

  print(f'Nome: {n} - Idade: {i}')



```

```python
#Etapa 01 B

import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(18,10))
plt.bar(nomes,idades)
plt.title('Alunos por Idade')
plt.ylabel('Idade')
```
